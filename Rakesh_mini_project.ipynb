```python
# ---------------- FUNCTIONS -------------------

def welcome_screen():
    print("-----------------------------------------")
    print("     STUDENT PROGRESS TRACKER SYSTEM")
    print("-----------------------------------------")

def get_student_details():
    name = input("Enter Student Name: ")
    clas = input("Enter Class        : ")
    roll = input("Enter Roll Number  : ")
    return name, clas, roll

def get_marks():
    n = int(input("How many subjects? "))
    marks = [int(input(f"Enter marks for Subject {i+1}: ")) for i in range(n)]
    return marks

def calculate_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    elif avg >= 35:
        return "D"
    else:
        return "Fail"

def check_result(marks):
    return "FAIL" if any(m < 35 for m in marks) else "PASS"

def display_report(name, clas, roll, total, avg, grade, result):
    print("\n-----------------------------------------")
    print("          STUDENT PROGRESS REPORT")
    print("-----------------------------------------")
    print(f"Name        : {name}")
    print(f"Class       : {clas}")
    print(f"Roll No     : {roll}")
    print(f"Total Marks : {total}")
    print(f"Average     : {avg:.2f}")
    print(f"Grade       : {grade}")
    print(f"Result      : {result}")
    print("-----------------------------------------")

# ---------------- MAIN PROGRAM -------------------

welcome_screen()
name, clas, roll = get_student_details()

while True:
    marks = get_marks()
    total = sum(marks)
    avg = total / len(marks)
    grade = calculate_grade(avg)
    result = check_result(marks)

    display_report(name, clas, roll, total, avg, grade, result)

    print("\n1. Enter marks again")
    print("2. View report again")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        continue
    elif choice == "2":
        display_report(name, clas, roll, total, avg, grade, result)
    elif choice == "3":
        print("Thank you for using the system!")
        break
    else:
        print("Invalid choice!")
```
